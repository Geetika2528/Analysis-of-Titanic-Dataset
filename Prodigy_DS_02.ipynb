```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "C:\\Users\\ASUS\\Prodigy_InfoTech\\TITANIC\\train.csv"
titanic_data = pd.read_csv(file_path)
print(titanic_data.head())
print("\nDimensions of the Titanic dataset:", titanic_data.shape)
print("\nColumn names and data types:")
print(titanic_data.dtypes)
print("\nSummary statistics for numerical columns:")
print(titanic_data.describe())
print("\nMissing values:")
print(titanic_data.isnull().sum())
mean_age = titanic_data['Age'].mean()
titanic_data['Age'].fillna(mean_age, inplace=True)

mode_embarked = titanic_data['Embarked'].mode()[0]
titanic_data['Embarked'].fillna(mode_embarked, inplace=True)

titanic_data.drop(columns=['Cabin'], inplace=True)
sns.barplot(x='Sex', y='Survived', data=titanic_data, ci=None)
plt.title('Survival Rate by Gender')
plt.show()
sns.barplot(x='Pclass', y='Survived', data=titanic_data, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.show()
sns.histplot(titanic_data['Age'], bins=20, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
